```python
import pandas as pd  # For handling CSV and DataFrame
import matplotlib.pyplot as plt  # For plotting

# Read CSV into a DataFrame
file_path = 'Canonicalized_smiles_data.csv'  # Replace with your CSV file path
data = pd.read_csv(file_path)

# Example of the column names in the CSV, adjust as necessary
data_cols = data.columns.tolist()

# Generate KDE plots for specified columns and add vertical lines
for col in ["KIT", "KIT-D816V"]:
    filtered_columns = [c for c in data_cols if col in c]
    if filtered_columns:
        fig = data[filtered_columns].plot.kde()
        plt.axvline(x=75, linestyle="dotted", color="orange")
        plt.title(f'KDE Plot for columns containing "{col}"')
        plt.xlabel('Value')
        plt.ylabel('Density')
        plt.show()
    else:
        print(f'No columns found for "{col}"')
```
